# Data visualization 

To retrieve the training and test data for the models, we will use the handy Python package called `yfinance` ([docs](https://aroussi.com/post/python-yahoo-finance)) which retrieves historical asset price information from Yahoo Finance and stores the data in a Pandas dataframe. 

```python
import yfinance as yf

apple_price = yf.Ticker("AAPL")
```

```python
apple_price.info
```

Above is all of the information we can retrieve via `yfinance` for a particular asset. In this project, we will try to predict the price of the `AAPL` stock. Let's retrive the historical data and make a plot! 

```python
ap_history = apple_price.history(period="3y")
ap_history
```

### Plot the stock price over time

```python
ap_close_prices = list(ap_history['Close'])
```

```python
import matplotlib.pyplot as plt
plt.plot(ap_close_prices)
plt.ylabel('Price in $')
plt.xlabel('Time (days)')
plt.show()
```
